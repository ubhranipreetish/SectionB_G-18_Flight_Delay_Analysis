**Insight:**

- The cancellation reason (`CANCELLATION_CODE`) and specific delay reason columns (`DELAY_DUE_*`) are highly messy/nullable because they are only recorded when those specific events (cancellations or delays >= 15m) actually happen.
- Actual delay fields (`DEP_DELAY`, `ARR_DELAY`) are occasionally missing, mostly for cancelled or diverted flights.

- Dataset contains missing values in delay columns
